```python
from openai import OpenAI
from google.colab import userdata

client = OpenAI(
    base_url="https://openrouter.ai/api/v1",
    api_key=userdata.get('OPEN_ROUTER_KEY'),
)

response = client.chat.completions.create(
    model="nvidia/nemotron-3-nano-30b-a3b:free",
    messages=[
        {
            "role": "system",
            "content": "You are web developer assistant."
        },
        {
            "role": "user",
            "content": "Tell me about pug"
        }
    ],
)

print("РЕЗУЛЬТАТ:")
print(response.choices[0].message.content)
```

```python
from letta_client import Letta
from google.colab import userdata

# https://app.letta.com/projects/default-project (авторизуемся, на этой странице находим и ключ, и пример кода для старта)
client = Letta(api_key=userdata.get('LETTA_API_KEY'))

agent = client.agents.create(
    name="openrouter_test",
    memory_blocks=[
        {
            "label": "persona",
            "value": "I am running on OpenRouter via Letta Cloud."
        },
        {
            "label": "human",
            "value": "Name: User\nFirst interaction: Testing OpenRouter support"
        }
    ],
    model="openrouter/arcee-ai/trinity-mini:free"  # имя модели из openRouter (проверено на мне лично!!!)
)

print(f"Agent created: {agent.id}")
```
